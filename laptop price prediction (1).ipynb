```python
import mlflow
import mlflow.sklearn
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error, max_error
import numpy as np

# ==============================
# Common transformation
# ==============================
step1 = ColumnTransformer(transformers=[
    ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0,1,6,7,8])
], remainder='passthrough')

# MLflow experiment setup
mlflow.set_experiment("Laptop Price Prediction v3")

# Metric calculation helper
def calculate_metrics(y_true, y_pred):
    r2 = r2_score(y_true, y_pred)
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100
    max_err = max_error(y_true, y_pred)
    return r2, mae, mse, rmse, mape, max_err

# Auto-promotion logic
from mlflow.tracking import MlflowClient
client = MlflowClient()

def auto_promote(model_name, r2, mae):
    """Promote model to staging if metrics are good."""
    if r2 > 0.80 and mae < 0.25:
        latest_version = client.get_latest_versions(model_name, stages=["None"])[-1].version
        client.transition_model_version_stage(
            name=model_name,
            version=latest_version,
            stage="Staging",
            archive_existing_versions=True
        )
        print(f"✅ Model promoted to Staging (version {latest_version})")
    else:
        print("⚠️ Model did not meet promotion criteria.")
```

```python
import time

def auto_promote(model_name, r2, mae):
    """Promote model to Staging if metrics are good and model is registered."""
    try:
        if r2 > 0.80 and mae < 0.25:
            # Wait briefly for model registration to complete
            time.sleep(2)

            # Get the latest model version safely
            versions = client.search_model_versions(f"name='{model_name}'")
            if not versions:
                print("⚠️ No model versions found for promotion.")
                return
            
            latest_version = max(int(v.version) for v in versions)

            # Promote to staging
            client.transition_model_version_stage(
                name=model_name,
                version=latest_version,
                stage="Staging",
                archive_existing_versions=True
            )
            print(f"✅ Model promoted to Staging (version {latest_version})")
        else:
            print("⚠️ Model did not meet promotion criteria.")
    except Exception as e:
        print("❌ Auto-promotion failed:", e)
```

```python
from sklearn.linear_model import Ridge

model_name = "Ridge_Regression_Model"

with mlflow.start_run(run_name="Ridge Regression"):
    # Model pipeline
    step2 = Ridge(alpha=1.0, random_state=3)
    pipe = Pipeline([
        ('step1', step1),
        ('step2', step2)
    ])

    # Train
    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_test)

    # Metrics
    r2, mae, mse, rmse, mape, max_err = calculate_metrics(y_test, y_pred)

    # Log parameters + metrics
    mlflow.log_param("alpha", 1.0)
    mlflow.log_metrics({
        "R2": r2,
        "MAE": mae,
        "MSE": mse,
        "RMSE": rmse,
        "MAPE": mape,
        "Max Error": max_err
    })

    # Log and register model
    mlflow.sklearn.log_model(pipe, name=model_name, registered_model_name=model_name)

    print(f"R2: {r2:.4f}, MAE: {mae:.4f}, MSE: {mse:.4f}, RMSE: {rmse:.4f}, MAPE: {mape:.2f}%, Max Error: {max_err:.4f}")
    print("✅ Ridge model logged successfully.")

    # Auto-promotion
    auto_promote(model_name, r2, mae)
```

```python
with mlflow.start_run(run_name="Ridge_Regression_Model"):
    # Fit model
    pipe.fit(X_train, y_train)
    
    # Predictions
    y_pred = pipe.predict(X_test)

    # Metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    
    print("R2 Score:", r2)
    print("MAE:", mae)

    # ✅ Log parameters, metrics, and model
    mlflow.log_param("model_type", "Ridge")
    mlflow.log_param("alpha", 10)
    mlflow.log_metric("R2_score", r2)
    mlflow.log_metric("MAE", mae)

    # ✅ Log model with signature
    mlflow.sklearn.log_model(
        pipe, 
        name="ridge_regression_model",
        input_example=X_train.head(1)
    )

print("✅ Ridge Regression model training & MLflow logging completed.")










```

```python
with mlflow.start_run(run_name="Lasso_alpha_0.001"):
    # Step 1: Preprocessing (OneHotEncode categorical columns)
    step1 = ColumnTransformer(transformers=[
        ('col_tnf', OneHotEncoder(sparse_output=False, drop='first'), [0, 1, 7, 8])
    ], remainder='passthrough')

    # Step 2: Model
    step2 = Lasso(alpha=0.001)

    # Combine steps into a pipeline
    pipe = Pipeline([
        ('step1', step1),
        ('step2', step2)
    ])

    # Train model
    pipe.fit(X_train, y_train)

    # Predict
    y_pred = pipe.predict(X_test)

    # Metrics
    r2 = r2_score(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)

    print(f"R2 Score: {r2}")
    print(f"MAE: {mae}")

    # Log parameters & metrics
    mlflow.log_param("model", "Lasso Regression")
    mlflow.log_param("alpha", 0.001)
    mlflow.log_metric("r2_score", r2)
    mlflow.log_metric("mae", mae)

    # Log the trained model
    mlflow.sklearn.log_model(
        sk_model=pipe,
        artifact_path="lasso_model",
        input_example=X_train.iloc[:1],
        registered_model_name="Lasso_Regression_Model"
    )

print("✅ Lasso model training and MLflow logging completed.")
```
